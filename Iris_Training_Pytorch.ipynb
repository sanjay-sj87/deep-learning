```python
from sklearn.datasets import load_iris
```

```python
import numpy as np
import pandas as pd
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

# Data

```python
dt = load_iris()
```

```python
dt.keys()
```

```python
X, Y = dt.data, dt.target
```

```python
dt.target_names
```

# Model Architecture

```python
model = nn.Sequential(
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    # nn.Softmax(dim=1)
)
```

```python
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
# model(X)
```

```python
# torch.argmax(model(X), dim=1)
```

```python
lossfn = nn.CrossEntropyLoss()
optSgd = optim.SGD(model.parameters(), lr=0.01)
```

# Training

```python

for _ in range(200):
    optSgd.zero_grad()
    Yp = model(X)
    loss = lossfn(Yp, Y)
    loss.backward()
    optSgd.step()

    print(loss.item())
```

# Inference

```python
Ynewp = model(X)
```

```python
sftout = torch.softmax(Ynewp, dim=1)
```

```python
torch.argmax(sftout, dim=1)
```

# Using Batches

```python
from torch.utils.data import DataLoader
```

```python
df = pd.DataFrame(X)
```

```python
df['Y'] = Y
```

```python
data = torch.tensor(df.to_numpy())
```

```python
modl = nn.Sequential(
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    # nn.Softmax(dim=1)
)
```

```python
lossfn = nn.CrossEntropyLoss()
batchSgd = optim.SGD(modl.parameters(), lr=0.01)
```

```python
batches = DataLoader(data, batch_size=10, shuffle=True)
```

```python
modl.train()
for epoch in range(200):
    batch_loss = 0
    for i, batch in enumerate(batches):
        batch_X = batch[:, :-1].float()
        batch_Y = batch[:, -1].long()

        batchSgd.zero_grad()
        batch_Yp = modl(batch_X)
        loss = lossfn(batch_Yp, batch_Y)
        loss.backward()
        batchSgd.step()

        batch_loss = batch_loss + loss.item()
    print(f"Epoch {epoch+1}: Total Loss = {batch_loss}")
```
